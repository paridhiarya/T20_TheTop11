# T20 DATA - Cleaning, Transformation and Preprocessing

## Importing required libraries

```python
import pandas as pd
import json
import re
```

## Processing File I : Match Summary

```python
with open('T20_Match_Results.json') as fptr:
    data = json.load(fptr)
data
```

### Convert the dictionary into a dataframe

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
```

```python
df_match.head()
```

### Analyzing the structure of the dataframe

```python
df_match.shape
```

### Renaming columns and cleaning data

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
df_match['ground'] = df_match['ground'].apply(lambda x: re.sub(r'\([^)]*\)', '', x))
df_match.head()
```

### Creating Dictionary to map 'team1 vs team 2' to 'match_id'
This is done in order to create enable us to ctreate a relationship between the two tables further down the project.

```python
match_ids_dict = {}
for index, row in df_match.iterrows():
    key1 = row["team1"] + ' Vs ' + row["team2"]
    key2 = row["team2"] + ' Vs ' + row["team1"]
    
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]
match_ids_dict
```

### Export the finished '.csv' file

```python
df_match.to_csv('clean_match_summary.csv', index = False)
```

## Processing File II : Batting Summary

### Convert the dictionary into a dataframe

```python
with open('T20_Batting_Summary.json') as fptr:
    data = json.load(fptr)
    all_records = []
    for rec in data:
            all_records.extend(rec['battingSummary'])
df_batting = pd.DataFrame(all_records)
```

```python
df_batting.head()
```

### Analyzing the structure of the dataframe

```python
df_batting.shape
```

```python
df_batting.info()
```

### Renaming columns and cleaning data

#### We convert the dismissal column into out/not out

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if x != x else "not_out")
```

```python
df_batting.head(15)
```

```python
df_batting.drop(columns = ['dismissal'], inplace = True)
df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('(c)', ''))
df_batting.head(15)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head()
```

### Export the finished '.csv' file

```python
df_batting.to_csv('clean_batting_summary.csv', index = False)
```

## Processing File IV : Player Info Summary
